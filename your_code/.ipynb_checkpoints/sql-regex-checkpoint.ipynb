```python
# How many distinct (different) actors' last names are there?

df = pd.read_sql_query("""

SELECT COUNT(DISTINCT last_name) AS 'Actors last name'

FROM actor""", engine)

df
```

```python
# In how many different languages where the films originally produced? (Use the column language_id from the film table)

df = pd.read_sql_query("""

SELECT COUNT(language_id) AS 'nr. of languages'
FROM film
""", engine)

df
```

```python
# How many movies were released with "PG-13" rating?

df = pd.read_sql_query("""

SELECT COUNT(rating) AS 'nr. movies pg-13 rating'
FROM film
WHERE rating = 'PG-13'
""", engine)

df
```

```python
# Get the 10 longest movies from 2006.

df = pd.read_sql_query("""

SELECT title, length, release_year
	FROM film
		WHERE release_year = '2006'
	ORDER BY length DESC
    LIMIT 10
""", engine)

df
```

```python
# Show rental info with additional columns month and weekday. Get 20.

df = pd.read_sql_query("""

SELECT *, EXTRACT(MONTH FROM rental_date) AS 'month', WEEKDAY(rental_date) as 'weekday' 
FROM rental
LIMIT 20
""", engine)

df
```

```python
# Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.

df['day_type'] = df.weekday.apply(lambda x: 'weekend' if x in ('6','7') else 'workday')

df
```

```python
# How many rentals were in the last month of activity?

df = pd.read_sql_query("""

SELECT COUNT(rental_id) AS 'total rentals'
	FROM rental
		WHERE rental_date REGEXP '2006-02'
""", engine)

df
```

```python
# Get film ratings.

df = pd.read_sql_query("""

SELECT film_id, title, rating FROM film

""", engine)

df
```

```python
# Get release years.

df = pd.read_sql_query("""

SELECT film_id, title, release_year FROM film

""", engine)

df
```
